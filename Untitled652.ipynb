```python
import pandas as pd
import os
import json
import yaml

# ==============================
# 📂 FILE PATHS
# ==============================
input_file = r"C:\Users\NXTWAVE\Downloads\Predictive Healthcare Model\PCA0000_2011_MDDS.xls"
output_dir = r"C:\Users\NXTWAVE\Downloads\Predictive Healthcare Model"

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# ==============================
# 📥 LOAD DATA
# ==============================
print("Loading Excel file...")
df = pd.read_excel(input_file)

print("Dataset Loaded Successfully!")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# ==============================
# 🧹 BASIC CLEANING
# ==============================
# Remove completely empty rows
df.dropna(how='all', inplace=True)

# Fill missing values (optional strategy)
df.fillna(method='ffill', inplace=True)

# Convert column names to lowercase
df.columns = df.columns.str.lower().str.strip()

# ==============================
# 💾 SAVE AS PICKLE (.pkl)
# ==============================
pkl_path = os.path.join(output_dir, "dataset.pkl")
df.to_pickle(pkl_path)
print(f"Saved Pickle: {pkl_path}")

# ==============================
# 💾 SAVE AS HDF5 (.h5)
# ==============================
h5_path = os.path.join(output_dir, "dataset.h5")
df.to_hdf(h5_path, key='data', mode='w')
print(f"Saved HDF5: {h5_path}")

# ==============================
# 💾 SAVE AS JSON (.json)
# ==============================
json_path = os.path.join(output_dir, "dataset.json")

# Convert DataFrame to dictionary format
data_dict = df.to_dict(orient="records")

with open(json_path, "w", encoding="utf-8") as f:
    json.dump(data_dict, f, indent=4)

print(f"Saved JSON: {json_path}")

# ==============================
# 💾 SAVE AS YAML (.yaml)
# ==============================
yaml_path = os.path.join(output_dir, "dataset.yaml")

with open(yaml_path, "w", encoding="utf-8") as f:
    yaml.dump(data_dict, f, allow_unicode=True)

print(f"Saved YAML: {yaml_path}")

# ==============================
# 🎯 DONE
# ==============================
print("\nAll formats saved successfully!")
```
